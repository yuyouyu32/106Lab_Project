```python
import pandas as pd
from sklearn.model_selection import train_test_split

data = pd.read_csv('./27+16_data.csv')
target_names = ['d_of_Bandwidth','Bandwidth','d_of_minRL','F_of_minRL','minRL', 'Class']
data
```

## All Data

```python
for target_name in target_names:
    if target_name == 'Class':
        continue
    target = data.loc[:, target_name]
    features = data.drop(columns=target_names)
    split_data = pd.concat([target, features],axis=1)
    split_data.to_csv(f'./Data/{target_name}.dat', sep=' ', index=True)
```

## Train and Test

```python
for target_name in target_names:
    target = data.loc[:, target_name]
    if target_name == 'Class':
        continue
    features = data.drop(columns=target_names)
    Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size=0.3,random_state=32)
    train_data = pd.concat([Ytrain, Xtrain],axis=1)
    train_data.reset_index(drop=True).to_csv(f'./Data/{target_name}_train.dat', sep=' ', index=True)
    test_data = pd.concat([Ytest, Xtest],axis=1)
    test_data.reset_index(drop=True).to_csv(f'./Data/{target_name}_test.dat', sep=' ', index=True)
```

## R2 Calculate

```python
from sklearn.svm import SVR
for target_name in target_names:
    target = data.loc[:, target_name]
    features = data.drop(columns=target_names)
    Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size=0.3,random_state=32)
    model = SVR(gamma='scale', kernel="rbf", C=100)
    model.fit(Xtrain, Ytrain)
    print(f'[SVR] [{target_name}] r2 = {model.score(Xtest, Ytest)}')
```

## Covariate Shift

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split

data = pd.read_csv('./27+16_data.csv')
target = data.loc[:, 'Class'].values
features = data.drop(columns=['d_of_Bandwidth','Bandwidth','d_of_minRL','F_of_minRL','minRL', 'Class']).values
# features = StandardScaler().fit_transform(features)
Xtrain, Xtest, Ytrain, Ytest = train_test_split(features,target,test_size=0.3,random_state=32)
```

```python
import matplotlib.pyplot as plt
Xtrain_v, Xtest_v, Ytrain_v, Ytest_v = train_test_split(features, target, test_size=0.3,random_state=32)
fig = plt.figure()
plt.rcParams['savefig.dpi'] = 1500       # 
plt.rcParams['figure.dpi'] = 1500        # 
ax = fig.add_subplot(121, projection='3d')
ax.scatter(xs=Xtrain_v[:, 0], ys=Xtrain_v[:, 1], zs=Xtrain_v[:, 2], marker='o', c=Ytrain_v, cmap='coolwarm')
ax_t = fig.add_subplot(122, projection='3d')
ax_t.scatter(xs=Xtest_v[:, 0], ys=Xtest_v[:, 1], zs=Xtest_v[:, 2], marker='*', c=Ytest_v, cmap='coolwarm')
plt.show()
```

```python
from sklearn.svm import SVC
Kernel = ["linear",'poly',"rbf","sigmoid"]
for kernel in Kernel:
    clf= SVC(kernel = kernel
             , gamma="auto"
             , degree = 1
             , class_weight = 'balanced'
            ).fit(Xtrain,Ytrain)
    print("The accuracy under kernel %s is %f" % (kernel,clf.score(Xtest,Ytest)))
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV
gamma_range = np.logspace(-10,1,50)
coef0_range = np.linspace(0,5,10)
param_grid = dict(gamma = gamma_range
                  ,coef0 = coef0_range)

cv = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=420)
grid = GridSearchCV(SVC(kernel = "linear",degree=1, class_weight = 'balanced')
                    , param_grid=param_grid
                    , cv=cv)
grid.fit(features, target)
print("The best parameters are %s with a score of %0.5f" % (grid.best_params_, grid.best_score_))
```

```python
from sklearn.metrics import classification_report

model = SVC(kernel='linear', C=1.0, gamma=grid.best_params_['gamma'], probability=True, class_weight = 'balanced')
model.fit(Xtrain, Ytrain)
result_report = classification_report(Ytest, model.predict(Xtest))
# model.predict_proba(Xtest)
print(result_report)
```

```python
from sklearn.metrics import matthews_corrcoef

matthews_corrcoef(Ytest, model.predict(Xtest)) 
```

```python
from sklearn.metrics import roc_auc_score as AUC
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(Ytest, model.decision_function(Xtest), pos_label=1, drop_intermediate=True) # return false positive rate, true positive rate and threshold
```

```python
area = AUC(Ytest, model.decision_function(Xtest))
print(area)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(fpr, tpr, color='red'
         ,label='ROC curve (area = %0.2f)' % area)
plt.plot([0, 1], [0, 1], color='black', linestyle='--')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
plt.show()
```
